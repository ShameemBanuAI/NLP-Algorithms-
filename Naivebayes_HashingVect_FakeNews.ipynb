# Importing necessary libraries

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer, HashingVectorizer
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
```

```python
df=pd.read_csv("news.csv",index_col=None)
df.head(5)
```

```python
dataset=df.drop("Unnamed: 0",axis=1)
dataset
```

# Input and Output Split

```python
x=dataset["text"]
y=dataset["label"]
```

# Train and Test set split

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.33,random_state=53)
```

# Applying Hashing Vector

```python
hash_vectorizer = HashingVectorizer(stop_words='english')
hash_train = hash_vectorizer.fit_transform(x_train)
print(hash_train)
hash_test = hash_vectorizer.transform(x_test)
print(hash_test)
```

# Model Creation

```python
from sklearn.naive_bayes import MultinomialNB
mul_clf=MultinomialNB()
mul_clf.fit(hash_train,y_train)
```

```python
from sklearn.naive_bayes import GaussianNB
gau_clf=GaussianNB()
gau_clf.fit(hash_train,y_train)
```

```python
from sklearn.naive_bayes import BernoulliNB
ber_clf=BernoulliNB()
ber_clf.fit(hash_train,y_train)
```

# Prediction

```python
pred=ber_clf.predict(hash_test)
```

# Evaluation Metrics

```python
score=accuracy_score(y_test,pred)
report=classification_report(y_test,pred)
```

```python
print("Accuracy : %0.3f " % score )
```
